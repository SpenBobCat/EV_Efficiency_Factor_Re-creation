```python
EV_EF_Range_data = [    {"Range": "0-100", "Efficiency Factor2": 3.77},    
                    {"Range": "101-240", "Efficiency Factor2": 3.54},    
                    {"Range": "241-300", "Efficiency Factor2": 3.43}]     
```

 - Created a list of dictionary objects, each representing a different range of electric vehicles, where each dictionary has two key-value pairs: "Range" and "Efficiency Factor2". 
 - The "Range" key maps to a string representing the range of the electric vehicle in miles, while the "Efficiency Factor2" key maps to a float value representing the efficiency factor of the electric vehicle in that range. 
 - The data represents a summary of the electric vehicles' efficiency in 3 different ranges: 0-100, 101-240, and 241-300 miles with an "Efficiency Factor": 3.77, 3.54, and 3.43 respectively.

```python
df = df.assign(Efficiency_Factor = None)
```

 - Have a new column named **"Efficiency_Factor"** is being added to the DataFrame df with all the values set to None.

```python
print(df.columns)
```

```python
efficiency_factors = {(0,100):3.77, (101,240):3.54, (241,300):3.43}
for index, row in df.iterrows():
    range_miles = row["Range"]
    if row["Efficiency_Factor"] is None:
        for key in efficiency_factors:
            if range_miles>= key[0] and range_miles<= key[1]:
                df.at[index, "Efficiency_Factor"] = efficiency_factors[key] 
```

 - Created a dictionary called "efficiency_factors" which contains the efficiency factors for three different ranges of electric vehicles.

 - Used the **iterrows()** function to iterate over each row of the DataFrame df. 
  - For each row, it retrieves the value of the "Range" column and assigns it to a variable named "range_miles". 
  - It then checks if the "Efficiency_Factor" value of the same row is None. 
  - If it is None, the code enters a for loop which iterates over the keys of the "efficiency_factors" dictionary. 

Inside the for loop, it checks if the "range_miles" variable is greater than or equal to the first element of the key tuple and less than or equal to the second element of the key tuple. If the condition is true, it uses the **at()** function to set the value of the "Efficiency_Factor" column of the current row to the value corresponding to the current key in the "efficiency_factors" dictionary.
